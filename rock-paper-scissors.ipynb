```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

from random import choice 

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.

option = ['stone', 'paper', 'scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
max_games = 5

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
won_2_win = 3

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

def comp_choice(): 
    return choice(option)

# = comp_choice()
#print(a)

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

def player_choice():
    user = input("Write your choice (stone/paper/scissors): ")
    user = user.lower();
    while (user != "stone" and user != "paper" and user != "scissors"):
        print(user);
        user = input("That choice is not valid. Enter your choice (stone/paper/scissors): ");
        user = user.lower();
    return user 

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins

def combat(user_points, computer_points, won_2_win): 
    if computer_points == user_points: 
        combat_result = 0
    elif computer_points >= won_2_win:
        combat_result = 1
    elif user_points >= won_2_win: 
        combat_result = 2
    return combat_result

def combat_winner(result_game): 
    if result_game == 0:
        print('0. There is a TIE! Play again')
    elif result_game == 1: 
        print('1. I am sorry. The computer has won')
    elif result_game == 2: 
        print('2. Congratulations. You have won! (: ')
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
 
def game_status(won_2_win, max_games): 
# Create two variables that accumulate the wins of each participant
    current_game = 0 
    acc_points_comp = 0 
    acc_points_user = 0

# Create a loop that iterates while no player reaches the minimum of wins
    while acc_points_comp < won_2_win and acc_points_user < won_2_win and current_game < max_games: 
        
# necessary to win. Inside the loop solves the play of the 
# machine and ask the player's.
        print('Current game:', current_game)
        user_choice = player_choice()
        computer_choice = comp_choice()
        print('Computer choice:', computer_choice)
#Compare them and update the value of the variables that accumulate the wins of each participant.

        if computer_choice == user_choice: 
            pass
        elif computer_choice == 'stone': 
            if user_choice == 'scissors': 
                acc_points_comp +=1
               
            elif user_choice == 'paper': 
                acc_points_user +=1
                
        elif computer_choice == 'paper': 
            if user_choice == 'stone': 
                acc_points_comp += 1
              
            elif user_choice == 'scissors': 
                acc_points_user += 1
               
        elif computer_choice == 'scissors': 
            if user_choice == 'stone': 
                acc_points_user +=1
            elif user_choice == 'paper': 
                acc_points_comp += 1
        else: 
            print('Wrong input. Try again!')
            
        current_game += 1
        print('User points:', acc_points_user)
        print('Computer points:', acc_points_comp)
     # Print by console the winner of the game based on who has more accumulated win
        
    combat_winner = combat(acc_points_user, acc_points_comp, won_2_win)   
    
    return combat_winner

 

winner = game_status(won_2_win, max_games)
combat_winner(winner)
```
